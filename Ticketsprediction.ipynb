# Modellash qismi

```python
X = df.drop('price',axis = 1)
y = df.price.copy()

X_test = df_test.copy()
```

```python
X_num = X.select_dtypes(include=['int64', 'float64'])
X_cat = X.select_dtypes(include=['object'])

num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
])

num_attribs = list(X_num)
cat_attribs = X_cat.columns

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])
```

```python
X_prepared = full_pipeline.fit_transform(X)
X_prepared[0]
```

```python
# Convert sparse matrix to dense array
dense_array = X_prepared[0].toarray()

# Print the dense representation
print(dense_array)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LogisticRegression
```

# LR Model

```python
lr_model = LinearRegression()
lr_model.fit(X_prepared, y)

ridge = Ridge()
ridge.fit(X_prepared, y)

lasso = Lasso()
lasso.fit(X_prepared, y)

rf_model = RandomForestRegressor()
rf_model.fit(X_prepared, y)

gb_model = GradientBoostingRegressor()
gb_model.fit(X_prepared, y)

dt_model = DecisionTreeRegressor()
dt_model.fit(X_prepared, y)

svm_model = SVR()
svm_model.fit(X_prepared, y)

log_model = LogisticRegression()
log_model.fit(X_prepared, y)
```

```python
# lr model

X_test_prepared = full_pipeline.transform(X_test)
y_pred = lr_model.predict(X_test_prepared)

scores = cross_val_score(lr_model, X_prepared, y, cv=5, scoring='neg_mean_squared_error')

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(np.sqrt(-scores))
```
